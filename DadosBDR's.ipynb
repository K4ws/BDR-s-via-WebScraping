```python
# Importações das Bibliotecas

import pandas as pd
from bs4 import BeautifulSoup
from bs4 import BeautifulSoup as soup
import requests
```

```python
# Site usado para pegar os BDR's é o InvestNews

url = 'https://investnews.com.br/financas/veja-a-lista-completa-dos-bdrs-disponiveis-para-pessoas-fisicas-na-b3/'

agent = {"User-Agent":"Mozilla/5.0"}

page = requests.get(url, headers=agent)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Identificar o Html da Tabela
html = soup.find('table', class_= 'has-background')
```

```python
# Passando o Html para um DataFrame usando o Pandas
df = pd.read_html(str(html))[0]
df
```

```python
# Definir a primeira linha como cabeçalho
df.columns = df.iloc[0]

# Remover a primeira linha do conteúdo
df = df.drop(df.index[0])
```
